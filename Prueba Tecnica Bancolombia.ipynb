# Poca correlacion con la variable objetivo Amount no se puede usar la regresion lineal , se va a usar el KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold,cross_val_score, cross_val_predict
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
```

```python
X_sample, _, y_sample, _ = train_test_split(X, Y, train_size=0.01, random_state=42)
```

```python
# Entrenar
kfold=KFold(n_splits=10,random_state=7,shuffle=True)
model=KNeighborsRegressor(n_neighbors=19,algorithm='kd_tree',weights='distance')
scoring='neg_mean_squared_error'
results= cross_val_score(model,X_sample,y_sample,cv=kfold,scoring=scoring)
y_pred = cross_val_predict(model, X_sample, y_sample, cv=kfold)
print(f"MSE:{results.mean()}")
```

```python
mse = mean_squared_error(y_sample, y_pred)
mae = mean_absolute_error(y_sample, y_pred)
r2 = r2_score(y_sample, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {mse**0.5:.2f}")
print(f"R Score: {r2:.3f}")
```

# NO FUE EFICIENTE USAR EL KKN ( 55.81 se equivoca en promedio), SE VA A USAR RANDOM FOREST REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor
kfold=KFold(random_state=7,n_splits=10,shuffle=True)
model=DecisionTreeRegressor(criterion='squared_error')
scoring='neg_mean_squared_error'
results= cross_val_score(model,X_sample,y_sample,cv=kfold,scoring=scoring)
y_pred = cross_val_predict(model, X_sample, y_sample, cv=kfold)
print(f"MSE:{results.mean()}")
```

```python
mae = mean_absolute_error(y_sample, y_pred)
mse = mean_squared_error(y_sample, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_sample, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R Score: {r2:.3f}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_sample, y=y_pred, alpha=0.4)
plt.plot([y_sample.min(), y_sample.max()], [y_sample.min(), y_sample.max()], 'r--')
plt.xlabel("Valor real")
plt.ylabel("Prediccin")
plt.title("Prediccin vs Real (DecisionTreeRegressor)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

El grfico de dispersin entre las predicciones del modelo y los valores reales muestra que la mayora de los puntos estn alineados con la diagonal, lo demuestra que el modelo realiza predicciones cercanas a la realidad. Esto representa las metricas obtenidas (MAE  7, RMSE  24, R  0.83) y confirma que el modelo de regresin basado en rboles de decisin es efectivo para predecir el monto de gasto por categora y cliente.
